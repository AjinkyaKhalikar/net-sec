```python
def unique_values(df, columns):

    for column_name in columns:
        print(f"Column: {column_name}\n{'-'*30}")
        unique_vals = df[column_name].unique()
        value_counts = df[column_name].value_counts()
        print(f"Unique Values ({len(unique_vals)}): {unique_vals}\n")
        print(f"Value Counts:\n{value_counts}\n{'='*40}\n")

col_names = df.select_dtypes(include='object').columns
unique_values(df,col_names)
```

```python
columns = ['duration', 'protocol_type', 'service', 'flag', 'src_bytes',
       'dst_bytes', 'wrong_fragment', 'hot', 'logged_in', 'is_guest_login',
       'count', 'srv_count', 'serror_rate', 'srv_serror_rate', 'rerror_rate']

x_train = x_train[columns]
x_test = x_test[columns]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
logistic = LogisticRegression(random_state = 42)
random_forest = RandomForestClassifier(n_estimators=100 , random_state=42,n_jobs=-1)
xgb = XGBClassifier(random_state = 42)
```

```python
logistic= logistic.fit(x_train,y_train)
```

```python
random_forest= random_forest.fit(x_train,y_train)
```

```python
xgb= xgb.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

def model_evaluator(model, x_train= x_train, y_train= y_train, x_test= x_test, y_test= y_test):
    
    y_train_pred = model.predict(x_train)
    y_pred = model.predict(x_test)

    print("Train_Set")
    print("accuracy score : ",accuracy_score(y_train, y_train_pred))
    print("Confusion Matrix: ",confusion_matrix(y_train, y_train_pred))
    print("classification report",classification_report(y_train, y_train_pred))
    print()

    print("Test_Set")
    print("accuracy score : ",accuracy_score(y_test, y_pred))
    print("Confusion Matrix: ",confusion_matrix(y_test, y_pred))
    print("classification report",classification_report(y_test, y_pred))
```

```python
print("Evaluation for Logistic Rergression")
model_evaluator(logistic)
```

```python
print("Evaluation for Random Forest")
model_evaluator(random_forest)
```

```python
print("Evaluation for xgboost")
model_evaluator(xgb)
```

```python
#Creating this function to get proper output either attacked or normal

def output(model, x):
    y_pred = model.predict(x)
    
    if y_pred == 0:
        return 'Normal'

    else:
        return 'Attacked' 

x_new = x_test[:1]
```

```python
output_rdf = output(random_forest,x_new)
print("output for random forest: ", output_rdf)

output_xgb = output(xgb, x_new)
print("output for xgboost classifier: ",output_xgb)
```

```python
output_logistic = output(logistic,x_new)
print("output for logistic: ",output_logistic)
output_ann = output(model, x_new)
print("output for ANN: ",output_ann)
```

# Lets go one step further in this step we will give output as attack type not binary(attack/normal) but what kind of attack it is or is it normal

```python
df.head()
```

```python
DoS_attacks   = ['neptune','teardrop','smurf','pod','back','land']
Probe_attacks = ['ipsweep','portsweep','nmap','satan']
R2L_attacks   = ['warezclient','guess_passwd','ftp_write','multihop','imap','warezmaster','phf','spy']
U2R_attacks   = ['buffer_overflow','rootkit','loadmodule','perl']

attack_type = []

for action in df['attack']:
    if action in DoS_attacks:
        attack_type.append('DoS')
    elif action in Probe_attacks:
        attack_type.append('Probe')
    elif action in R2L_attacks:
        attack_type.append('R2L')
    elif action in U2R_attacks:
        attack_type.append('U2R')
    else:
        attack_type.append('Normal')

print(attack_type)
```

```python
df['attack_type'] = attack_type

df.head()
```
